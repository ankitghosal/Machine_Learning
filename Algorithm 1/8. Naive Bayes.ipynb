# Naive Bayes

```python
import pandas as pd
import matplotlib.pyplot as plt
import  seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r'D:\Project Data Set\cgpa_score-placed\cgpa_score-placed.csv')
dataset.head(3)
```

```python
# check if the graph is normal distribution
plt.figure(figsize=(5,3))
sns.kdeplot(data=dataset['resume_score'])
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='cgpa', y='resume_score', data= dataset, hue='placed')
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset['placed']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

# Gausian NB

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
gnb.score(x_test,y_test) , gnb.score(x_train,y_train)
```

```python
plt.figure(figsize=(5,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(), clf=gnb)
plt.show()
```

### Almost a perfect curve

# Multinomial NB

```python
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
```

```python
mnb.score(x_test,y_test) , mnb.score(x_train,y_train)
```

```python
plt.figure(figsize=(5,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(), clf=mnb)
plt.show()
```

#### Not a perfect curve atall.

# Bernoulli NB

```python
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
```

```python
bnb.score(x_test,y_test) , bnb.score(x_train,y_train)
```

```python
plt.figure(figsize=(5,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(), clf=bnb)
plt.show()
```

### Not showing any decision region atall. as 50-50 chance
