# L1 & L2 Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('D:\Project Data Set\Melbourne_housing_FULL.csv')
dataset.head()
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
# remove the unnecessary columns
col_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 'Distance',
             'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
dataset = dataset[col_to_use]
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
# Fill some columns with 0
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
dataset.isna().sum()
```

```python
# for Landsize & BuildingArea - we'll put the mean value
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())

dataset.isna().sum()
```

```python
# rest values will be dropped
# price is not tailored bcz it'll give wrong info during training. So dropping it.
dataset.dropna(inplace = True)

dataset.isna().sum()
```

### Need to doOne Hot Encoding to convert Words into digits.

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset.head()
```

```python
x = dataset.drop('Price', axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x,y, train_size=0.3, random_state=2)
```

```python
# First use regular Linear Regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_x, train_y)
```

```python
reg.score(test_x, test_y)
```

```python
reg.score(train_x, train_y)
```

### Here train & test dataset showing different results. Better for Train model than test. So clearly Overfitting.
Neet to use Lasso regression to get rid of that.

## L1 Regularization

```python
# using L1 Regression
from sklearn import linear_model

lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)

lasso_reg.fit(train_x, train_y)
```

```python
lasso_reg.score(test_x, test_y)
```

```python
lasso_reg.score(train_x, train_y)
```

## L2 Regularization

```python
from sklearn.linear_model import Ridge

ridge_reg= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(train_x, train_y)
```

```python
ridge_reg.score(test_x, test_y)
```

```python
ridge_reg.score(train_x, train_y)
```
