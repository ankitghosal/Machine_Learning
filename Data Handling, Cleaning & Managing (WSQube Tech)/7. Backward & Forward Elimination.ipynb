```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from mlxtend.feature_selection import SequentialFeatureSelector   # a new library
```

```python
dataset = pd.read_csv(r"C:\\Users\\Ankit\\Downloads\\Forward_Elimination.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
x = dataset.iloc[:,:-1]   # all rows, -1 column
y = dataset['Outcome']
```

```python
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
fs = SequentialFeatureSelector(lr, k_features=3 ,forward= False)     
# here estimator is 'lr' (basically the sklearn module) ; x has 5 features(  it'll be K_feature)
```

### Forward = True (Forward Elimination)
### Forward = False (Backward Elimination)

### make sure before Fit
####  X string
####  X NaN (Blank)
####  x ONLY 0,1 ( not a lot of number on which it'll be done)

```python
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```
