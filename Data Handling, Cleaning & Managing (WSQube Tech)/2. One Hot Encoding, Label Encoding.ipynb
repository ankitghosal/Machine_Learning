# Ordinal Encoding
1. Sklearn     2. MAP function

```python
df = pd.DataFrame({'Size': ['s','m','l','xl','s','m','l','s','s','xl','m']})
df.head(3)
```

```python
ord_data = [['s','m','l','xl']]
```

## 1. Sklean

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = ord_data)
oe.fit(df[['Size']])    # use [['Size']] not [] for Ordinal Encoding
```

```python
df['Size_en'] = oe.transform(df[['Size']])
```

```python
df
```

## 2. MAP function

```python
ord_data1 = {'s': 5, 'm':6, 'l':7, 'xl':8}    # here we can assign random numbers
```

```python
df['Size_en_map'] = df['Size'].map(ord_data1)
```

```python
df
```
