Feature Scaling -> 1. Standardization & 2. Normalization

# Standardization
Xnew = (Xi - Xmean)/(Std Deviation)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\\Users\\Ankit\\Downloads\\HousingDataSet_2.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset['AGE'].fillna(dataset['AGE'].mean(), inplace=True)
```

```python
plt.figure(figsize=(9,3))
sns.distplot(dataset['AGE'])
plt.show()
```

```python
dataset.describe()
```

# Standarddization Process

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[['AGE']])
```

```python
dataset['AGE_ss'] = pd.DataFrame(ss.transform(dataset[['AGE']]),columns=['X'])
```

```python
dataset.head(4)
```

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset['AGE'])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset['AGE_ss'])
plt.show()
```

# Normalization or Min Max Scaler

Xnew = (Xi - Xmin)/(Xmax - Xmin)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(dataset[['AGE']])
```

```python
dataset['AGE_ms'] = ms.transform(dataset[['AGE']])
dataset.head(4)
```

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
plt.title('Before')
sns.distplot(dataset['AGE'])

plt.subplot(1,2,2)
plt.title('After')
sns.distplot(dataset['AGE_ms'])

plt.show()
```
