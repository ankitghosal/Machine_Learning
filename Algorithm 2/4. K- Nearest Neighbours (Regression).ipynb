# K- Nearest Neighbours (Regression)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r'D:\Project Data Set\multiple_linear_regression_dataset.csv')
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
x = dataset.drop(columns='income')
y = dataset['income']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor(n_neighbors=2)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
knn.score(x_train,y_train)
```

```python
for i in range(1,30):
    knn1 = KNeighborsRegressor(n_neighbors=i)
    knn1.fit(x_train,y_train)
    print(i,knn1.score(x_train,y_train),knn1.score(x_test,y_test))
```
