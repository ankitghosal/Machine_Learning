# 11. DBSCAN Clustering Algorithm
#### When we have 'Non-Linear Dataset'

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
import warnings
warnings.filterwarnings('ignore')
```

```python
# importing Non-Linear dataset from sklearn library
from sklearn.datasets import  make_moons
```

```python
x,y = make_moons(n_samples=250, noise=0.05)
```

```python
# Creating a dictionary for better dealing of Data
# df = {'data1':x[:,0] , 'data2':x[:,1], 'output':y }  # removing output:y. only keeping data1, data2
df = {'data1':x[:,0] , 'data2':x[:,1]}
```

```python
dataset = pd.DataFrame(df)
dataset.head(3)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='data1', y='data2', data=dataset)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.5, min_samples=5)
dataset['Predict'] = db.fit_predict(dataset)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='data1', y='data2', data=dataset, hue='Predict')
plt.show()
```
