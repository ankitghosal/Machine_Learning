# 16. Bagging(Meta-Estimator, Random Forest) [Classification]

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
import warnings
warnings.filterwarnings('ignore')
import numpy as np
```

```python
from sklearn.datasets import make_moons
x,y = make_moons(n_samples=1000, noise = 0.2)  # putting noise=0.2 to gget more noise.(0.05 was giving pretty good data)
```

```python
df = {'x1':x[:,0], 'x2':x[:,1], 'y':y}
```

```python
dataset = pd.DataFrame(df)
```

```python
sns.scatterplot(x='x1',y='x2', data=dataset, hue='y')
```

```python
x_a = dataset.iloc[:,:-1]
y_a = dataset['y']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x_a,y_a, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

## Bagging Classifier

```python
bg = BaggingClassifier(estimator=SVC(), n_estimators=30)
bg.fit(x_train, y_train)
bg.score(x_train,y_train), bg.score(x_test, y_test)
```

```python
sv = SVC()
sv.fit(x_train, y_train)
sv.score(x_train,y_train), sv.score(x_test, y_test)
```

### Random Forest

```python
rf = RandomForestClassifier(n_estimators=30)
rf.fit(x_train, y_train)
rf.score(x_train, y_train), rf.score(x_test, y_test)
```
