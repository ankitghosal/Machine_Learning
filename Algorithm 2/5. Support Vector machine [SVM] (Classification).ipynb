# Support Vector machine [SVM] (Classification)
Can be used for both Linear & Non-Linear Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r'D:\Project Data Set\cgpa_score_placed\cgpa_score_placed.csv')
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='cgpa', y='resume_score', data=dataset, hue='placed')
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset['placed']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=15)
```

```python
# Our data is of Classification type. So we'll choose SVC
from sklearn.svm import SVC
```

```python
sv = SVC(kernel='linear') # using Linear kernel as the map looks like linearly seperable
sv.fit(x_train, y_train)
```

```python
sv.score(x_test, y_test)*100
```

```python
# Checking Overfitting
sv.score(x_train, y_train)*100
```

### Seeing the Graph

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=sv)
plt.show()
```

####  Not  properly Linearly sperable. As some points are falling  in  different regions  & score also not coming as desired.
