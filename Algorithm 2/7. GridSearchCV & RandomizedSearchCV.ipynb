# 7. GridSearchCV & RandomizedSearchCV

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r"D:\Project Data Set\Salary\Level_salary.csv")
dataset.head(3)
```

```python
x = dataset.iloc[:,:-1]
y = dataset['Salary']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor # as I know this dataset is Regression type
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
dt.score(x_train, y_train) , dt.score(x_test, y_test)
```

Good Fitting. only 1% Overfitting noticed. - to remove even that we use GridSearchCV & RandomizedSearch CV.

## GridSearchCV use

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
# Need to create a dictinary for DecisionTreeRegressor elements so I can use them inside GridSearchCV
df = {"criterion": ["squared_error", "friedman_mse", "absolute_error", "poisson"],
    "splitter": ["best", "random"],
    "max_depth": [i for i in range (2,20)]}
```

```python
gd = GridSearchCV(DecisionTreeRegressor(), param_grid=df)
gd.fit(x_train, y_train)
```

```python
gd.best_params_     # It provides Best Parameters to overcome overfitting.
```

```python
gd.best_score_
```

```python
dt = DecisionTreeRegressor(criterion='squared_error', max_depth=7, splitter= 'best')
dt.fit(x_train,y_train)
```

```python
dt.score(x_train, y_train) , dt.score(x_test, y_test)
```

# RandomizedSearchCV use

```python
rd = RandomizedSearchCV(DecisionTreeRegressor(), param_distributions=df, n_iter=20)
rd.fit(x_train, y_train)
```

```python
rd.best_params_    # randomly  searches and finds out best parameters. Everytime we run the fitting best parameter data change.
```

```python
rd.best_score_
```

```python
dt = DecisionTreeRegressor(criterion='poisson', max_depth=12, splitter= 'best')
dt.fit(x_train,y_train)
```

```python
dt.score(x_train, y_train) , dt.score(x_test, y_test)
```
