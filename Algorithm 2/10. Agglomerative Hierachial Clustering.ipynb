# 10. Agglomerative Hierachial Clustering

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv(r'D:\Project Data Set\New_Iris.csv')
dataset.head(3)
```

##### Check if Data is Linearly Seperable. Agglomeartive Clustering only works on Linearly seperable Data.

```python
# To check iff the data is Linearly Seperable - Create a Pair Plot
plt.figure(figsize=(4,3))
sns.pairplot(data= dataset)
plt.show()
```

###  Create Dendogram

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(10,5))
sc.dendrogram(sc.linkage(dataset,method='single', metric='euclidean'))  # Each method gives different outcome. But fundamentally same.
plt.savefig('Dendogram.jpg')
plt.show()
```

#### Here There'll be 2 clusters (Chek from saved image in File 'Dendogram'.) 

### Find Cluster

```python
from sklearn.cluster import AgglomerativeClustering

ac = AgglomerativeClustering(n_clusters=2, linkage='single')   # n_clulster = 2 as we found 2 clusters in the diagram
dataset['Predict'] = ac.fit_predict(dataset)
```

```python
dataset.head(2)
```

#### Now again check pairplot to see the clustering

```python
plt.figure(figsize=(3,2))
sns.pairplot(data=dataset, hue='Predict')
plt.show()
```
