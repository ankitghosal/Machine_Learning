```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
import warnings
warnings.filterwarnings('ignore')
import numpy as np
```

```python
dataset = pd.read_csv(r'D:\Project Data Set\groceries_trimmed_data.csv')
dataset.head(3)
```

```python
dataset.shape
```

```python
# Need to remove NaN from the dataset and make a fresh list
market = []

for i in range(0, dataset.shape[0]):
    cus = []
    for j in dataset.columns:
        if type(dataset[j][i])==str:
            cus.append(dataset[j][i])

    market.append(cus)
```

```python
# to fit market in a single list
l = []
for i in market:
    for j in i:
        l.append(j)
```

```python
import collections
```

```python
p = collections.Counter(l)
```

```python
# makng a dictionary of it p
d = {"Item Name":p.keys(), "values":p.values()}
```

```python
# making a dataframe of dict d
pd.DataFrame(d).sort_values(by=["values"], ascending=False)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr = TransactionEncoder()
tr.fit(market)
```

```python
df = pd.DataFrame(tr.transform(market), columns=tr.columns_)
```

# Using FP  Growth Algo now:

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
fpgrowth(df, min_support=0.05, use_colnames=True, max_len=3).sort_values(by=['support'])
```
