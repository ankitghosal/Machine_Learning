# K- Nearest Neighbours

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r'D:\Project Data Set\Social_Network_Ads\New_Social_Network_Ads.csv')
dataset.head(3)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='Age',y='EstimatedSalary', data= dataset, hue= 'Purchased')
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x = dataset.iloc[:,:-1]
y = dataset['Purchased']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler().fit(x)
x =pd.DataFrame(sc.transform(x), columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
# p = 1 -> Manhatten dis
# p = 2 -> Euclidean dis (default)
```

```python
knn.score(x_test,y_test)
```

```python
knn.score(x_train,y_train)
```

```python
knn.predict([[1.083596, -0.990844]])
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=knn)
plt.show()
```

## Finding Right Value of Neighbor

```python
for i in range(1,30):
    knn1 = KNeighborsClassifier(n_neighbors=i)
    knn1.fit(x_train,y_train)
    print(i,knn1.score(x_train,y_train),knn1.score(x_test,y_test))

# 3  - comming as a perfect Neighbor
```
